### Which country has won the most men's gold medals in singles badminton over the years?(Sort the results alphabetically by the player's name)


```python
badminton_dataframe = oo[oo.Sport.str.contains('badminton', case=False)]
badminton_dataframe
```

```python
singles_gold_mens_badminton_dataframe = badminton_dataframe[(badminton_dataframe.Gender == 'Men')&(badminton_dataframe.Medal == 'Gold')&(badminton_dataframe.Event == 'singles')]
singles_gold_mens_badminton_dataframe
```

### Which three countries have won the most medals in recent years(from 1984 and 2008)?


```python
country_with_most_Medals = oo.City.value_counts(ascending=False)
#type(country_with_most_Medals)
country_with_most_Medals
```

### Display the male gold medal winners for the 100m track and field sprint event over the years
