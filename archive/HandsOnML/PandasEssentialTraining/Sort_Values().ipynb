# Sort_Values()

```python
import pandas as pd
oo = pd.read_csv('../../datasets/PandasEssentialTraining/olympics.csv',skiprows=4)
oo.head()



oo.Athlete.sort_values()
```

```python
#The sort_values method has a parameter which have inplace=False, which means that the a new series will be returned.
#If we want to capture these changes we will need to assign the result to a new series
ath = oo.Athlete.sort_values()
```

```python
#Lets sort by the Edition of the olympics and the Athletes name
oo.sort_values(by=['Edition', 'Athlete'])
```

```python
type(oo.sort_values(by=['Edition', 'Athlete']))
```
