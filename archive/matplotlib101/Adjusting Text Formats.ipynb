# Text Format for rc parameters

```python
import matplotlib
```

```python
matplotlib.rcParams['font.size'] = 20
matplotlib.rc('font', size=20)
matplotlib.rc('font', family='sans-serif') 
```

```python
matplotlib.rcParams.update({'font.size':10, 'font.family':'serif'})
```

# Text Format for x-y labels

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3]
y = [2,4,6]

fontparams = {'size':16, 'fontweight':'light', 'family':'monospace','style':'normal'}


plt.plot(x,y)

plt.xlabel('xlabel', size=20, fontweight='semibold', family='serif', style='italic')
plt.ylabel('ylabel', fontparams)

plt.show()
```

```python
#Font manager
from matplotlib.font_manager import findSystemFonts
findSystemFonts(fontpaths=None, fontext='ttf')
```

```python
#Latext support
import numpy as np

#prepare a curve of square numbers 
x = np.linspace(0,200,100)
y1 = x #prepare an array of y equals to x squared
y2 = x + 20

#plot a curve of square numbers 
plt.plot(x,y1,label='$x$')
plt.plot(x,y2,label='$x^2+\alpha$')
plt.legend()
plt.show()
```
