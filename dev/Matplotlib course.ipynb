```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd

%reload_ext autoreload
%autoreload 2
```

# Functional plot

```python
x_1 = np.linspace(0,5,10)
y_1 = x_1**2
plt.plot(x_1, y_1)
plt.title('Days squared chart')
plt.xlabel('Days')
plt.ylabel('Days Squared')
```

# Multiple plots

```python
plt.subplot(2,2,1)
plt.plot(x_1, y_1, 'purple')
plt.subplot(2,2,2)
plt.plot(x_1, y_1, 'g')
plt.subplot(2,2,3)
plt.plot(x_1, y_1, 'pink')
plt.subplot(2,2,4)
plt.plot(x_1, y_1, 'black')
```

# Print Multiple Plots
    A figure is an object that is going to contain all the plot elements, and it can contain many axis

```python
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.figure.html
fig_1 = plt.figure(figsize=(5,4), # figsize = Width, height in inches.
                   dpi=100)  #dpi= The resolution of the figure in dots-per-inch.
axes_1 = fig_1.add_axes([0.1, 0.1, 0.9, 0.9])
axes_1.set_xlabel('Days')
axes_1.set_ylabel('Days Squared')
axes_1.set_title('Dyas squared chart')
axes_1.plot(x_1, y_1, label='x/x^2')
axes_1.plot(y_1, x_1, label='x^2/x')
axes_1.legend(loc=0)

# Create another axis inside this axis
axes_2 = fig_1.add_axes([0.45, 0.45, 0.4, 0.3])
axes_2.set_xlabel('Days')
axes_2.set_ylabel('Days Squared')
axes_2.set_title('Dyas squared chart')
axes_2.plot(x_1, y_1, 'r')
axes_2.text(0,40, 'This is a text message')
```

# Subplots

```python
fig_2, axes_2 = plt.subplots(figsize=(8,4), nrows=1, ncols=3)
#space between plots
plt.tight_layout()
axes_2[0].set_title('Plot 2')
axes_2[0].set_xlabel('X')
axes_2[0].set_ylabel('X Squared')
axes_2[0].plot(x_1, y_1)
```
