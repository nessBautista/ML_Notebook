```python
import pandas as pd 
pd.set_option('display.max_rows', None)
oo = pd.read_csv('../../datasets/PandasEssentialTraining/olympics.csv', skiprows= 4)
oo.head()
```

# In which events did Jesse Owens win a medal?

```python
#evaluate where the Athlete name contains owens, then Using Boolean indexing get a new data frame to get data
#from owens
owen_series = oo.Athlete.str.contains('owens', case=False)
owen_dataframe = oo[owen_series]
owen_dataframe
```

```python
#we can cleary see the data frame, but we just want the event column. We can use value_counts 
owen_events = owen_dataframe.Event.value_counts()
owen_events
```

# Which country has won the most men's gold medals in badminton over the years?(Sort the results alphabetically by the player's name)

```python
#Lets obtain the indices that evalutes true for: sport== 'badminton', Medal=='Gold', Gender='Men'
boolean_indexes = (oo.Sport.str.contains('badminton', case=False)) & (oo.Medal.str.contains('Gold', case=False))&(boolean_indexes)
response_dataframe = (oo[boolean_indexes])
response_dataframe
```

```python
#lets sort this dataframe by Athlete's name
sorted_response_dataframe = response_dataframe.sort_values(by=['Athlete'])
sorted_response_dataframe
```

```python
#Finally let get the country with more Medals
response_dataframe_more_medals = sorted_response_dataframe.NOC.value_counts()
response_dataframe_more_medals
```

# Which three countries have won the most medals in recent years(from 1984 to 2008)


```python
#First get the indices by time
response_indices = (oo.Edition >= 1985)&(oo.Edition <=2008)
#then get a new data frame
recent_olympics_dataframe = oo[response_indices]
recent_olympics_dataframe
```

```python
#now count countries
recent_olympics_dataframe.NOC.value_counts()
```

```python
#now to specify the top three
recent_olympics_dataframe.NOC.value_counts().head(3)
```

# Display the male gold medal winners for the 100m Track & Field sprint event over the years.
# List the result starting with most recent. Show the olympic city, edition, athlete and country they represent.

```python
GoldMedal_Men_100mEvent = oo[(oo.Gender == 'Men') & (oo.Event == '100m') & (oo.Medal == 'Gold')]
GoldMedal_Men_100mEvent
```

```python
#Now we need to sort by the most recent
sorted_GoldMedal_Men_100mEvent = GoldMedal_Men_100mEvent.sort_values(by=['Edition'], ascending=False)
sorted_GoldMedal_Men_100mEvent
```

```python
#finally we need to create a data frame with only: olympic city, edition, athlete and country they represent
subframe_goldMedal_Men_100mEvent = sorted_GoldMedal_Men_100mEvent[['City', 'Edition', 'Athlete', 'NOC']]
```
