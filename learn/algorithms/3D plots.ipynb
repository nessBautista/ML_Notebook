```python
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import matplotlib
import numpy as np
%matplotlib notebook

```

```python
# Plot a Vector
fig1 = plt.figure(1, figsize=(5,5))
ax1 = fig1.add_subplot(projection='3d')
ax1.set_xlim([-1,1])
ax1.set_ylim([-1,1])
ax1.set_zlim([-1,1])
ax1.set_xlabel("x axis")
ax1.set_ylabel("y axis")
ax1.set_zlabel("z axis")

def plot_vectors(origins, 
                 vectors, 
                 axe):
    for (o,v) in zip(origins,vectors):
        
        axe.quiver(o[0],o[1],o[2],
                    v[0],v[1],v[2])        

    axe.view_init(0,0)    


o = [0,0,0]
a = [1,1,1]
b = [-1,1,1]
c = [1,1,-1]
plot_vectors(origins = [o, o, o], vectors = [a,b,c], axe = ax1)
```
