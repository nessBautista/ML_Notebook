```python
import numpy as np
import matplotlib.pyplot as plt 

%matplotlib inline
```

```python
evens= np.arange(2,11,2,dtype=np.int32)
```

### When ploting a single array, matplotlib infers that data is in the X axis 

```python
plt.plot(evens)
```

```python
#plt.plot(x,y)
plt.plot(evens,evens**2)
```

```python
#Adding the 'label' parameter
plt.plot(evens,evens**2,label='x^2')
plt.legend()
#plt.savefig('output.png') #supported formats: pdf, png, svg, post
#plt.savefig('output1.png',dpi=200) #dpi=dot per inch, it is the resolution
#recommended resolutions at: https://docs.microsoft.com/en-us/office/troubleshoot/powerpoint/change-export-slide-resolution
plt.show()
```
