```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(3,2))
X= np.linspace(-np.pi,np.pi,300)
C,S= np.cos(X), np.sin(X)

plt.plot(X,C)
plt.plot(X,S)

plt.show
```
